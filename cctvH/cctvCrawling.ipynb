```python
import pandas as pd
import os

b = [0,0]

for file in os.listdir('./cctv_csv'):
    a = pd.read_csv('./cctv_csv/'+file)
    if os.path.exists('./cctv_csv/'+file):
        os.remove('./cctv_csv/'+file)
    a.drop('Unnamed: 0',axis=1,inplace=True)
    for i,col in enumerate(a.columns):
        b[i] = a[col].sort_values(ascending =True).reset_index(drop=True)
    
    cd = pd.concat([b[0],b[1]],axis = 1)
    cd.to_csv('./cctv_csv/'+file)
print('끝')
```
